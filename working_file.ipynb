## Description

**Is the SARIMA model useful for stock prediction? If so, when is it useful?**

This GitHub project utilized the yfinance library to import historical stock data for Apple Inc. (AAPL), NVIDIA (NVDA), and Tesla (TSLA). These stocks were chosen due to their differing behaviors: historically, AAPL is considered a stable stock, while NVDA and TSLA are more volatile. Using this data, SARIMA models were developed to predict stock prices for stocks with varying levels of volatility.
 
 **SARIMA Model: Seasonal Autoregressive Integrated Moving Average (SARIMA) model is a time series forecasting method implemented to capture the underlying patterns and trends in stock price data.**

The SARIMA model provides a robust framework for predicting stock prices by capturing seasonal patterns and cyclic behaviors inherent in sequential data. Its foundation in statistical time series analysis lends credibility to the predictions. The model's interpretable parameters offer deeper insights into the dynamics driving stock price movements. However, it is important to note that SARIMA has limitations in capturing external factors such as news events or sudden market shocks, which may impact stock prices but are not directly incorporated into the model. Despite this, SARIMA remains a valuable tool for predicting stock prices based on historical patterns and internal data dynamics.

## Objective
The objective of this project is to determine whether a robust SARIMA model can accurately predict a company's stock price using historical data and market trends. The primary goal is to use predictive insights to inform strategic investment decisions aimed at maximizing shareholder profit and optimizing portfolio performance. The first step was to identify how much historical data is necessary to train the machine learning models for the most accurate predictions, using specific metrics for comparison (see section below). The second step was to assess whether this tool is accurate enough for real-world application by analyzing the potential financial gains or losses.

For AAPL, NVDA, and TSLA, historical data for the following time ranges was downloaded to predict stock prices for the period from 06-01-2024 to 06-11-2024:

## Metrics Used for Comparison
To determine which historical data time range is best for each stock, the following metrics were used for comparison:
1. **P-value**: The p-value measures the probability that the observed results (or something more extreme) could have occurred under the null hypothesis.
2. **R-squared value**: The R-squared value (the coefficient of determination), represents the proportion of variance in the dependent variable that is predictable from the independent variable(s).
3. **Root Mean Squared Error (RMSE)**: The RMSE measures the average magnitude of the errors between predicted and actual values. It is the square root of the average of the squared differences between predictions and actual observations.
4. **Mean Squared Error (MSE)**: The MSE is the average of the squares of the errors, which are the differences between the actual values and the predicted values.
5. **Akaike Information Criterion (AIC)**: The AIC value is used for model selection that balances model fit and complexity, estimating the relative quality of statistical models for a given set of data.
6. **Bayesian Information Criterion (BIC)**: The BIC value is use for model selection that also balances model fit and complexity, similar to AIC, but with a stronger penalty for models with more parameters.

## Functions


### SARIMA Model Functions

The sarima_forecast function generates forecasts using the SARIMA model and returns values with confidence intervals

```python

def sarima_forecast(model, df, forecast_date, Feature_Scaler):
    print("forecast \n")
    # Forecast
    n_periods = (datetime.strptime(forecast_date, '%Y-%m-%d') - df.index[-1]).days
    fitted, confint = model.predict(n_periods=n_periods, return_conf_int=True)
    index_of_fc = pd.date_range(df.index[-1] + pd.DateOffset(days=1), periods = n_periods, freq='B')
    
    # Ensure the forecast does not go beyond the desired forecast date
    if index_of_fc[-1] > pd.Timestamp(forecast_date):
        index_of_fc = index_of_fc[index_of_fc <= pd.Timestamp(forecast_date)]
        fitted = fitted[:len(index_of_fc)]
        confint = confint[:len(index_of_fc)]

    # Make series for plotting purpose
    fitted_series = pd.Series(fitted.values, index=index_of_fc)
    lower_series = pd.Series(confint[:, 0], index=index_of_fc)
    upper_series = pd.Series(confint[:, 1], index=index_of_fc)

    #Concatenate the original DataFrame with forecasted values and confidence intervals
    df_result = pd.concat([df, fitted_series, lower_series, upper_series], axis=1)
    df_result.columns = ["Actual", "Prediction", "Low", "High"]

    #Inverse transform the scaled values to their original scale
    for column in df_result.columns:
      df_result[column] = Feature_Scaler.inverse_transform(df_result[column].values.reshape(-1,1))

    return df_result
```

The process_model function fits the SARIMA model to a time series dataset using the pmdarima library 

```python
def process_model(stock_transformed):
    print("process model \n")
    
    # Ensure the index is a DateTimeIndex for easy filtering
    stock_transformed.index = pd.to_datetime(stock_transformed.index)

    # Filter out weekends
    stock_transformed = stock_transformed[stock_transformed.index.to_series().dt.dayofweek < 5]
    
    sarima_model = pm.auto_arima(stock_transformed["Close"], start_p=1, start_q=1,
                         test='adf',
                         max_p=3, max_q=3,
                         m=12, #12 is the frequency of the cycle (yearly)
                         start_P=0,
                         seasonal=True, #set to seasonal
                         d=None,
                         D=1, #order of the seasonal differencing
                         trace=False,
                         error_action='ignore',
                         suppress_warnings=True,
                         stepwise=True)
    
    return sarima_model
```

The print_metrics function calculates and prints several key metrics used to evaluate the performance of a stock prediction model.

```python
def print_metrics(df_result, model, stock_name):
    stock_ticker = yf.Ticker(stock_name)
    
    # import actual closing prices into real_values dataframe
    real_values = yf.download(stock_name, end='2024-06-12')
    
    # calculate AIC and BIC
    aic = model.aic()
    bic = model.bic()
    print(f"AIC: {aic}")
    print(f"BIC: {bic}")

    # Ensure indices match
    df_result.index = pd.to_datetime(df_result.index)
    real_values.index = pd.to_datetime(real_values.index)
    
    # Remove weekends from df_result
    df_result = df_result[df_result.index.weekday < 5]
    
    # calculate p-value using OLS regression
    X = sm.add_constant(df_result.loc['2024-06-01':'2024-06-11', "Prediction"])
    y = real_values.loc['2024-06-01':'2024-06-11', "Close"]
    
    # Ensure indices match after adding constant
    X = X.reindex(y.index)
    
    # fit OLS model (ols regression)
    ols_model = sm.OLS(y, X).fit()
    
    # extract p-value for the 'Prediction' variable
    p_value = ols_model.pvalues.iloc[1]  
    print(f"P-value: {p_value}")

    # Calculate R-squared
    r_squared = ols_model.rsquared
    print(f"R-squared: {r_squared}")
    
    
    # Calculate RMSE and MSE
    actual_values = real_values.loc['2024-06-01':'2024-06-11', 'Close']
    predicted_values = df_result.loc['2024-06-01':'2024-06-11', 'Prediction']
    mse = mean_squared_error(actual_values, predicted_values)
    rmse = np.sqrt(mse)
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")



    

    
```

The serialize_deserialize function writes the SARIMA model to a file using serialization and reads the model with deserialization

```python
def serialize_deserialize(sarima_model):
    print("serialize \n")
    # Serialize with Pickle and save it as pkl
    with open('sarima_model.pkl', 'wb') as pkl:
        pickle.dump(sarima_model, pkl)

    # Desiarilize the content of the file back into a Python object
    with open('sarima_model.pkl', 'rb') as pkl:
        loaded_model = pickle.load(pkl)

    return loaded_model
```

The sarima_load function calls all necessary functions to construct the model

```python
def sarima_load(stock_data, feature_scaler):
    # remove any duplicate index
    stock_data = stock_data.loc[~stock_data.index.duplicated(keep='first')]
    # filter only required data
    stock_data = stock_data[['Close']]
    # transform current stock data
    stock_transformed = pd.DataFrame(np.squeeze(feature_scaler.fit_transform(stock_data), axis=1), columns=["Close"], index=stock_data.index)
    # call function to fit a seasonal ARIMA model
    sarima_model = process_model(stock_transformed)
    # call function to serialize / deserialize SARIMA model with pickle
    sarima_model = serialize_deserialize(sarima_model)
    return stock_transformed, sarima_model
```

The create_plot function plots the predicted values

```python
def create_plot(test, stock_name):
    # plot
    plt.figure(figsize=(15,7))
    plt.plot(test["Actual"][-120:], color='#1f76b4')
    plt.plot(test["Prediction"], color='darkgreen')
    plt.fill_between(test.index,
                test["Low"],
                test["High"],
                color='k', alpha=.15)

    plt.title(f"SARIMA - Forecast of {stock_name} Stock Price")
    plt.show()
```

The a_vs_p function plots Actual vs. Predicted Stock Prices

```python
import matplotlib.pyplot as plt

def a_vs_p(df_result, stock_hist, stock_name):
    
    plt.figure(figsize=(10, 7))
    
    # Scatter plot for actual vs. predicted values
    plt.scatter(stock_hist["Close"], df_result["Prediction"][-7:], color='darkgreen', marker='o')
    
    # Plotting the line y = x for reference
    plt.plot([stock_hist["Close"].min(), stock_hist["Close"].max()],
             [stock_hist["Close"].min(), stock_hist["Close"].max()], 
             color='red', linestyle='--', label='y = x')
    
    # Adding title and labels
    plt.title(f"Actual vs. Predicted Stock Prices - {stock_name}")
    plt.xlabel('Actual Prices')
    plt.ylabel('Predicted Prices')
    plt.legend()
    
    # Show the plot
    plt.show()
```
